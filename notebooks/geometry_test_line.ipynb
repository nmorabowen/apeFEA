```python
from apeFEA.core import Node, Line
from apeFEA.core import FEMConfig
from apeFEA.geometry import Geometry

import matplotlib.pyplot as plt
```

```python
FEMConfig.set_dimension(dimension=2,
                        nDof=2)
```

```python
geo= Geometry(name='Line Test')
geo.nodes.create_node(
    coords=[0,0],
    id=1
)
geo.nodes.create_node(
    coords=[6,8],
    id=2
)
geo.nodes.create_node(
    coords=[16,8],
    id=3
)
geo.nodes.create_node(
    coords=[16,0],
    id=4
)

geo.lines.create_line(
    node_j=geo.nodes.get_node_by_id(1),
    node_k=geo.nodes.get_node_by_id(2),
    id=1
)
geo.lines.create_line(
    j_id=2,
    k_id=3,
    id=2
)
geo.lines.create_line(
    j_id=4,
    k_id=3,
    id=3
)

geo.lines.print_lines_info()
```

```python
fig, ax = plt.subplots(figsize=(8, 8))

geo.plot_nodes(ax=ax)
geo.plot_lines(ax=ax)
```
