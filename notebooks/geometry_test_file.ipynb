```python
from apeFEA.core import Node
from apeFEA.core import FEMConfig
from apeFEA.geometry import Geometry
```

```python
FEMConfig.set_dimension(
    dimension=2,
    nDof=2
)
```

```python
# Geometry Creation

geo=Geometry(name='Geometry No1')

geo.nodes.create_node(coords=[0,0],
                id=1)
geo.nodes.create_node(coords=[6,8],
                id=2)
geo.nodes.create_node(coords=[16,8],
                id=3)
geo.nodes.create_node(coords=[16,0],
                id=4)

geo.nodes.print_nodes_info()

geo.nodes._used_ids
```

```python
geo.plot_nodes(backend='matplotlib', 
               show=True)
```

```python
geo.plot_nodes(backend='plotly', 
               show=True)
```

```python
# Geometric modification

geo.nodes.modify_node(
    node=geo.nodes.get_node_by_id(3),
    new_coords=[16,16],
)

geo.nodes.print_nodes_info()
```
